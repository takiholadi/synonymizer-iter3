## Пример 2

```python
data_dir = os.path.join('..',  'crawler-advego', 'input', 'original')
text_path = os.path.join(data_dir, 'construction.txt')
with open(text_path, mode='r', encoding='utf-8') as f:
    text = f.read()
```

```python
frame, synonymized_text = mix_backtranslation(text, pipelines)
```

```python
print(synonymized_text)
```

Закидываю этот просинонимированный текст в Адвего онлайн-антиплагиат, получаю скор уникальности: **91%**. У оригинального текста было **0%**.

Просматриваю по абзацу, для сравнения:

```python
for original, synonymized in zip(text.split('\n\n'), synonymized_text.split('\n\n')):
    print('BEFORE:\n{}'.format(original))
    print('AFTER:\n{}'.format(synonymized))
    print('\n')
```
